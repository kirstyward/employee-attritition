## Exploratory Data Analysis

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from string import ascii_letters

sns.set(style="white")
```

```python
#Read in data
df= pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv', index_col='EmployeeNumber')
pd.set_option('display.max_columns', 40)
df.head()
```

```python
#Check for missing values
df[df.isnull().any(axis=1)]
```

```python
df.describe()
```

There are variables with no variation in them, shown by a standard deviation (std) of 0. These variables are not informative, so let's drop them from the dataframe.

```python
#Find variables with only one value
nunique = df.apply(pd.Series.nunique)

#Drop those variables
constant_cols = nunique[nunique == 1].index
df = df.drop(constant_cols, axis=1)

print("The columns {} had only one unique value and were dropped from the dataset.".format(list(constant_cols)))
```

```python
#Data Quality Checks
boolean_checks = [df['YearsInCurrentRole'] > df['YearsAtCompany'],
                  df['TotalWorkingYears'] < df['YearsAtCompany'],
                 df['YearsWithCurrManager'] > df['YearsAtCompany'],
                 df['YearsSinceLastPromotion'] > df['YearsAtCompany']]

for i in boolean_checks:
    print(any(i))
```

```python
# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

## Feature Engineering
Next, I am creating new features in the dataset that could help explain attrition:

- *ScaledSalaryHike* : How large was someone's salary increase compared to their performance? Calculated by dividing `PercentSalaryHike` by `PerformanceRating`
- *AverageTenure* : The average amount of time an employee spends at a workplace. Calculate by dividing `TotalWorkingYears` by `NumCompaniesWorked`

```python
#Feature Engineering - Performance Scaled Salary Hike
df['ScaledSalaryHike'] = df['PercentSalaryHike'] / df['PerformanceRating']

sns.distplot(df['ScaledSalaryHike'])
```

```python
sns.boxplot(x='ScaledSalaryHike',y='Attrition',data=df,palette='rainbow')
```

```python
#Feature Engineering - Average Number of Years at each company worked
df['AverageTenure'] = df['TotalWorkingYears'] / (df['NumCompaniesWorked'] + 1)

sns.distplot(df['AverageTenure'])
```

```python
sns.boxplot(x='AverageTenure',y='Attrition',data=df,palette='rainbow')
```
